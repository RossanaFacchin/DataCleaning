Verifico se le date sono tutte valide, se sono scritte con un formato diverso o se sono mancanti oppure Nan

```python
# 2. Estrai solo il "formato" di ogni data, ignorando i valori specifici
#    Esempio: "2023-05-14" -> "dddd-dd-dd"
#             "14/05/2023" -> "dd/dd/dddd"

import re
import pandas as pd

col_data = "EXAMDATE"  # sostituisci con il nome esatto della colonna date nel tuo dataset

def estrai_formato(valore):
    if pd.isna(valore):
        return "NaN"
    valore = str(valore)
    return re.sub(r'\d', 'd', valore)

formati = df[col_data].apply(estrai_formato)

print("\nFormati distinti trovati nella colonna:")
print(formati.value_counts())
```

```python
# 3. Prova a convertire in formato datetime e individua i valori problematici
date_convertite = pd.to_datetime(df[col_data], errors='coerce')

# Righe che NON sono state convertite correttamente (ma non erano già NaN in origine)
problemi = df[df[col_data].notna() & date_convertite.isna()]

print(f"\nNumero di valori data non convertibili: {len(problemi)}")
print(problemi[[col_data]])
```

Controllare la validità della collona AGE

```python
col_age = "AGE"

# 1. Valori mancanti
nulli = df[col_age].isna().sum()
totali = len(df)
percentuale = (nulli / totali) * 100
print(f"Valori mancanti in '{col_age}': {nulli}/{totali} ({percentuale:.2f}%)")

# 2. Statistiche descrittive di base (min, max, media, mediana, ecc.)
print(f"\nStatistiche descrittive di '{col_age}':")
print(df[col_age].describe())

# 3. Tipo di dato (deve essere numerico)
print(f"\nTipo di dato: {df[col_age].dtype}")

# 4. Controllo range plausibile (in ADNI i partecipanti sono adulti, tipicamente 55-95 anni)
eta_min_plausibile = 18   # nessun paziente dovrebbe avere meno di 18 anni
eta_max_plausibile = 110  # nessun paziente dovrebbe avere più di 110 anni

fuori_range = df[(df[col_age] < eta_min_plausibile) | (df[col_age] > eta_max_plausibile)]
print(f"\nRighe con età fuori range plausibile ({eta_min_plausibile}-{eta_max_plausibile}): {len(fuori_range)}")
print(fuori_range[[col_age]])

# 5. Valori negativi o pari a zero (chiaramente errati)
invalidi = df[df[col_age] <= 0]
print(f"\nValori di età <= 0: {len(invalidi)}")
print(invalidi[[col_age]])

# 6. Valori non numerici mascherati da stringhe (es. spazi vuoti, "NA" scritto come testo)
non_numerici = df[pd.to_numeric(df[col_age], errors='coerce').isna() & df[col_age].notna()]
print(f"\nValori non numerici in '{col_age}': {len(non_numerici)}")
print(non_numerici[[col_age]])
```

Conteggio duplicati colonna ID pazienti

```python
# Conteggio dei duplicati con un ciclo for sui valori unici
duplicati_rid = {}

for rid in df["RID"].unique():
    conteggio = (df["RID"] == rid).sum()
    if conteggio > 1:
        duplicati_rid[rid] = conteggio

print(f"Numero di RID duplicati (con più di una riga): {len(duplicati_rid)}")
for rid, count in duplicati_rid.items():
    print(f"RID {rid}: {count} righe")
```
