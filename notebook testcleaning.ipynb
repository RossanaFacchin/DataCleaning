Trovare i valori Nan o nulli per ogni colonna con un ciclo for

```python
for col in df.columns:
    num_null = df[col].isna().sum()
    perc_null = (num_null / len(df)) * 100

    print(f"\nColonna: {col}")
    print(f"Valori nulli: {num_null}")
    print(f"Percentuale nulli: {perc_null:.2f}%")
```

In ordine dal più sporco al più pulito

```python
null_summary = pd.DataFrame({
    'colonna': df.columns,
    'null_count': [df[col].isna().sum() for col in df.columns],
    'null_percent': [(df[col].isna().sum() / len(df)) * 100 for col in df.columns]
})

print(null_summary.sort_values('null_count', ascending=False))
```

Solo le colonne che superano il 40% di valori nulli

```python
count = 0

for col in df.columns:
    percentuale = df[col].isna().mean() * 100

    if percentuale > 40:
        print(f"{col}: {percentuale:.2f}% nulli")
        count += 1

print(f"\nNumero di colonne con più del 40% di valori nulli: {count}")
```
